#  Chi square testing -  fouls


```python
# Null Hypothesis (H0): There is no association between the obersved foul & yellow cards between home and away
# Alternative Hypothesis (H1): There is an association between the obersved fouls & yellow cards between home and away teams
```

```python
databritclean
```

```python
databritclean.dtypes
```

```python
totalaycs = databritclean['A_Yellow_Cards'].sum()
avgaycs = databritclean['A_Yellow_Cards'].mean()
totalhycs = databritclean['H_Yellow_Cards'].sum()
avghycs = databritclean['H_Yellow_Cards'].mean()
print('tot away yellow:', totalaycs)
print('avg away yellow:', avgaycs)
print('tot home yellow:', totalhycs)
print('avg home yellow:', avghycs)
print('tot yellows:', totalaycs + totalhycs)
print('avg yellows:', ((avgaycs+avghycs)/2))
```

```python
totalafs = databritclean['A_Fouls'].sum()
avgafs = databritclean['A_Fouls'].mean()
totalhfs = databritclean['H_Fouls'].sum()
avghfs = databritclean['H_Fouls'].mean()
print('tot away fouls:', totalafs)
print('avg away fouls:', avgafs)
print('tot home fouls:', totalhfs)
print('avg home fouls:', avghfs)
print('tot fouls:', totalafs + totalhfs)
print('avg yellows:', ((avgafs+avghfs)/2))
```

```python
st.ttest_ind(databritclean['H_Yellow_Cards'],databritclean['A_Yellow_Cards'], alternative='two-sided')
```
